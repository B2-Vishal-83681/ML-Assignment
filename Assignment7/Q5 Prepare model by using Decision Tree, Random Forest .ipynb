### split data into x and y 

```python
x = df[['hair','feathers','eggs','milk','aquatic','toothed','backbone','breathes','venomous','legs','tail','domestic','catsize']]
y = df['type']
```

### split data into train and test

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y,train_size=0.7,random_state=123456)
```

### create model

```python
def create_dt():
    from sklearn.tree import DecisionTreeClassifier

    model = DecisionTreeClassifier(max_depth=3)
    model.fit(x_train, y_train)
    return model 
```

```python
def create_rm():
    from sklearn.ensemble import RandomForestClassifier

    model = RandomForestClassifier()
    model.fit(x_train, y_train)
    return model
```

```python
models = [(create_dt(),'Descision Tree'),(create_rm(),'Random Forest')]
```

### eveluate model

```python
from sklearn.metrics import  accuracy_score, precision_score, recall_score, f1_score
def evaluate_model(model_1, model_name):
    y_pred = model_1.predict(x_test)
    y_true = y_test

    accuracy = accuracy_score(y_true, y_pred)

    precision = precision_score(y_true, y_pred, average='macro')

    recall = recall_score(y_true, y_pred, average='macro')

    f1 = f1_score(y_true, y_pred, average='macro')
 
    return model_name, accuracy, precision, recall, f1
```

```python
evaluate_rows = []
for (model_1, model_name) in models:
    evaluate_rows.append(evaluate_model(model_1, model_name))
    
df_result = pd.DataFrame(evaluate_rows, columns=['Algorithm','Accuracy','Precision','Recall','f1'])
df_result
```
