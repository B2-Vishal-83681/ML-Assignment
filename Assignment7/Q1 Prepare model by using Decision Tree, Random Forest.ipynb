```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sb
```

### load the dataset

```python
df = pd.read_csv('Churn.csv')
```

```python
from sklearn.preprocessing import LabelEncoder

encode_surname = LabelEncoder()
df['Surname'] = encode_surname.fit_transform(df['Surname'])

encode_geography = LabelEncoder()
df['Geography'] = encode_geography.fit_transform(df['Geography'])

encode_gender = LabelEncoder()
df['Gender'] = encode_gender.fit_transform(df['Gender'])
```

### split data into x and y

```python
x = df[['Geography','Gender','Age','Balance','IsActiveMember']]
y = df['Exited']
```

### split data into train and test dataset

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=12345)
```

### train the module

```python
def create_descision_tree():
    from sklearn.tree import DecisionTreeClassifier

    model = DecisionTreeClassifier(max_depth=3)
    model.fit(x_train, y_train)
    return model
```

```python
def create_random_forest():
    from sklearn.ensemble import RandomForestClassifier

    model = RandomForestClassifier()
    model.fit(x_train, y_train)
    return model 
```

```python
models = [(create_descision_tree(), 'Descision Tree', (2,0)),
          (create_random_forest(), 'Random Forest',(2,1))]
```

### model evaluate

```python
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score
def evaluate_model(model,model_name):
    y_pred = model.predict(x_test)
    y_true = y_test

    accuracy = accuracy_score(y_true, y_pred)

    precision = precision_score(y_true, y_pred)

    recall = recall_score(y_true, y_pred)

    f1 = f1_score(y_true, y_pred)

    return model_name, accuracy, precision, recall, f1
```

```python
evaluate_rows = []
for (model, model_name, position) in models:
    evaluate_rows.append(evaluate_model(model,model_name))

df_result = pd.DataFrame(evaluate_rows, columns=['Algorithm', 'Accuracy', 'Precision', 'Recall','f1'])
df_result
```
