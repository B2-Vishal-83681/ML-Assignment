## Data Pre-processing

```python
df.drop(['Unnamed: 0'], axis=1, inplace=True)
df.head()
```

## Create hierarchical cluster

```python
from scipy.cluster import hierarchy

linkage = hierarchy.linkage(df, 'single')
dendo = hierarchy.dendrogram(linkage)
```

```python
linkage = hierarchy.linkage(df, 'complete')
dendo = hierarchy.dendrogram(linkage)
```

```python
linkage = hierarchy.linkage(df, 'average')
dendo = hierarchy.dendrogram(linkage)
```

## Use ski-kit Learn

```python
from sklearn.cluster import AgglomerativeClustering

model = AgglomerativeClustering(n_clusters=4, linkage='complete')
model.fit(df)
```

```python
model.n_clusters_
```

```python
model.labels_
```

```python
figure = plt.figure()

colors = ['red','green','yellow','purple']
for k in range(len(colors)):
    plt.scatter(df['Qual_miles'][model.labels_== k], df['Bonus_miles'][model.labels_==k], color=colors[k])
    plt.xlabel('Qualifying Miles')
    plt.ylabel('Bonus Miles')
```

## create DBSCAN model

```python
from sklearn.cluster import DBSCAN
model = DBSCAN(eps=8.0, min_samples=4)
model.fit(df)
```

```python
model.labels_
```

```python
figure = plt.figure()

colors = ['red','green','pink','blue']
for k in range(len(colors)):
    plt.scatter(df['Qual_miles'][model.labels_==k], df['Bonus_miles'][model.labels_==k], color=colors[k])
    plt.xlabel('Qualifying Miles')
    plt.ylabel('Bonus Miles')
```
