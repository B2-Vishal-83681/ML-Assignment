```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

### 1 NewsPaper

```python
df = pd.read_csv('NewspaperData.csv')
df.head()
```

#### find the covariance between the columns to check if there is any relationship between them

```python
# using numpy
covariance = np.cov(df['daily'], df['sunday'])
covariance
```

#### find the correlation between the columns to check if what is the degree of relationship

```python
correlation = np.corrcoef(df['daily'], df['sunday'])
correlation
```

### split the data into dependenct and independent columns

```python
y = df['sunday']
x = df.drop(['sunday','Newspaper'], axis=1)
```

### create the model

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(x,y)
```

```python
# show the observed values
plt.scatter(x['daily'], y, label='Observed Values', color='green')

# draw mean line
mean_value = np.ones(y.size) * y.mean()
plt.plot(x['daily'], mean_value, color='red')
plt.scatter(x['daily'], mean_value, label='Mean Value', color='red')

# draw the regression line
plt.plot(x['daily'], model.predict(x), color='blue')
plt.scatter(x['daily'], model.predict(x), color='blue', label='Expected Value')

plt.xlabel('Daily')
plt.ylabel('Sunday')
plt.title('Regression Line Graph')
plt.legend()
```
