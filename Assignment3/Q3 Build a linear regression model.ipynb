```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
df = pd.read_csv('TvMarketing.csv')
df.head()
```

### split the data into dependent and independent columns

```python
y = df['Sales']
x = df.drop('Sales', axis=1)
```

### create the model

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(x,y)
```

```python
plt.scatter(x['TV'], y, label='Observed Values', color='red')

mean_value = np.ones(y.size) * y.mean()
plt.plot(x['TV'], mean_value, color='green')
plt.scatter(x['TV'], mean_value, label='Mean Value', color='green')

plt.plot(x['TV'], model.predict(x), color='blue')
plt.scatter(x['TV'], model.predict(x), label='Expected Values', color='blue')

plt.xlabel('TV')
plt.ylabel('Sales')
plt.title('Regression Line Graph')
plt.legend()
```
