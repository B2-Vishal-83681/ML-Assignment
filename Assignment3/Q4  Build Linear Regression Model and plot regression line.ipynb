```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
df = pd.read_csv('IceCreamData.csv')
df.head()
```

### split the data into dependent and independent columns

```python
y = df['Revenue']
x = df.drop('Revenue', axis=1)
```

### create the model

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(x,y)
```

```python
plt.scatter(x['Temperature'], y, label='Observed Value', color='red')

mean_value = np.ones(y.size) * y.mean()
plt.plot(x['Temperature'], mean_value, color='green')
plt.scatter(x['Temperature'], mean_value, label='Mean Value', color='green')

plt.plot(x['Temperature'], model.predict(x), color='blue')
plt.scatter(x['Temperature'], model.predict(x), label='Expected Value', color='blue')

plt.xlabel('Temperature')
plt.ylabel('Revenue')
plt.title('Regression Line Graph')
plt.legend()
```
